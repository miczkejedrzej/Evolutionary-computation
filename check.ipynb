```python
import os  
import glob

import csv

# Open and read the CSV file
path = "./assignment_3/results"
files = glob.glob(os.path.join(path, "*.csv"))
for file in files:
    with open(file,mode="r") as f:
        reader = csv.reader(f)
        for row in reader:
            seq2 = [int(element) for element in row[0].split(";")]
            unique_count2 = len(set(seq2))
            print(file)
            print(unique_count2)
```
